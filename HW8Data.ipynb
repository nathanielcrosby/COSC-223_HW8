```python
import pandas
import numpy as np
import matplotlib.pyplot as plt

data = pandas.read_csv("QueueData.csv")
```

```python
data = data.to_numpy()
```

```python
theoretical_N = [(data[i,0]*(0.5) / (0.5 - data[i,0])) for i in range(len(data[:,0]))]

theoretical_T = [((1 - data[i,0]) / (0.5 - data[i,0])) for i in range(len(data[:,0]))]

theoretical_T
```

```python
plt.figure(figsize=(16,8))
plt.title('E[N] vs. p', fontsize=20)

plt.plot(data[:,0], data[:,1], 'k', color = 'red', label="Emperimental")
plt.plot(data[:,0], theoretical_N, 'k', color = 'blue', label="Theoretical")

plt.xlabel('p values (Geometric probability to determine arrival times)', fontsize=15)
plt.legend(loc="upper left", fontsize=12)
plt.ylabel('Average Number of Jobs in Server (E[N])', fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(16,8))
plt.title('E[T] vs. p', fontsize=20)

plt.plot(data[:,0], data[:,2], 'k', color = 'red', label="Emperimental")
plt.plot(data[:,0], theoretical_T, 'k', color = 'blue', label="Theoretical")

plt.xlabel('p values (Geometric probability to determine arrival times)', fontsize=15)
plt.legend(loc="upper left", fontsize=12)
plt.ylabel('Average Job Response Time (E[T])', fontsize=15)
plt.show()
```
